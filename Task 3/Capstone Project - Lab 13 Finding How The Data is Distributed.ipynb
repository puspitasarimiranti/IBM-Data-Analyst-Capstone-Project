### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
#check missing data
print('\nValue counts for missing data in LanguageHaveWorkedWith column:\n')
print( missing_data['LanguageHaveWorkedWith'].value_counts())

print('\nValue counts for missing data in LanguageWantToWorkWith column:\n')
print( missing_data['LanguageWantToWorkWith'].value_counts())
```

```python
#check unique value
print('\nThere are', df['LanguageHaveWorkedWith'].nunique(), 'unique LanguageHaveWorkedWith Total values in the survey:')
print(df['LanguageHaveWorkedWith'].value_counts())

print('\nThere are', df['LanguageWantToWorkWith'].nunique(), 'unique LanguageWantToWorkWith Total values in the survey:')
print(df['LanguageWantToWorkWith'].value_counts())
```

```python
#impute (replace) all the empty rows in the column with the majority value
df['LanguageHaveWorkedWith'].replace(np.nan, "HTML/CSS;JavaScript;TypeScript", inplace=True)
df['LanguageWantToWorkWith'].replace(np.nan, "Python", inplace=True)
```

```python
#check unique value again
print('\nThere are', df['LanguageHaveWorkedWith'].nunique(), 'unique LanguageHaveWorkedWith Total values in the survey:')
print(df['LanguageHaveWorkedWith'].value_counts())

print('\nThere are', df['LanguageWantToWorkWith'].nunique(), 'unique LanguageWantToWorkWith Total values in the survey:')
print(df['LanguageWantToWorkWith'].value_counts())
```

```python
#MODEL 1A

#initialize an empty list
all_languages = []

#iterate through the column, handling missing values with .dropna()
#assuming languages are separated by a semicolon ';' as in the original code
for languages_str in df['LanguageHaveWorkedWith'].dropna():
    
    #split the string by ';' and strip whitespace from each language name
    all_languages.extend([lang.strip() for lang in languages_str.split(';')])

#count language occurrences using pandas.Series.value_counts()
language_counts_series = pd.Series(all_languages).value_counts()

#convert the Series to a DataFrame for easier plotting/display
language_counts_haveworked_df = pd.DataFrame(language_counts_series)

#rename the column to something more descriptive (optional)
language_counts_haveworked_df.columns = ['CountHaveWorked']

print(language_counts_haveworked_df)
```

```python
#MODEL 1B

#initialize an empty list
all_languages_wantwork = []

#iterate through the column, handling missing values with .dropna()
#assuming languages are separated by a semicolon ';' as in the original code
for languages_str_wantwork in df['LanguageWantToWorkWith'].dropna():
    
    #split the string by ';' and strip whitespace from each language name
    all_languages_wantwork.extend([lang.strip() for lang in languages_str_wantwork.split(';')])

#count language occurrences using pandas.Series.value_counts()
language_counts_series_wantwork = pd.Series(all_languages_wantwork).value_counts()

#convert the Series to a DataFrame for easier plotting/display
language_counts_wantwork_df = pd.DataFrame(language_counts_series_wantwork)

#rename the column to something more descriptive (optional)
language_counts_wantwork_df.columns = ['CountWantWork']

print(language_counts_wantwork_df)
```

```python
#MODEL 2
from collections import Counter

# Function to count language frequencies
def count_languages(series):
    all_languages_2 = []
    for entry in series.dropna():
        all_languages_2.extend(entry.split(';'))
    return Counter(all_languages_2)

# Get frequencies for both categories
worked_with_counts = count_languages(df['LanguageHaveWorkedWith'])
want_to_work_with_counts = count_languages(df['LanguageWantToWorkWith'])

# Convert to DataFrames for easier manipulation
df_worked = pd.DataFrame(worked_with_counts.items(), columns=['Language', 'WorkedWithCount'])
df_want = pd.DataFrame(want_to_work_with_counts.items(), columns=['Language', 'WantToWorkWithCount'])

# Merge the two DataFrames
merged_df = pd.merge(df_worked, df_want, on='Language', how='outer').fillna(0)

print(df_worked)
print(df_want)
print(merged_df)
```

```python
#create Venn diagram
!pip install matplotlib_venn
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
```

```python
#get sets of languages
set_worked = set(worked_with_counts.keys())
set_want = set(want_to_work_with_counts.keys())

#example for two sets (Worked With and Want to Work With)
plt.figure(figsize=(8, 8))
venn2([set_worked, set_want], ('Worked With', 'Want to Work With'))
plt.title('Overlap of Programming Languages')
plt.show()
```

```python
#prepare data for grouped bar chart
plot_df = merged_df.melt(id_vars='Language', var_name='Category', value_name='Count')

plt.figure(figsize=(12, 7))
sns.barplot(x='Language', y='Count', hue='Category', data=plot_df)
plt.title('Frequency of Programming Languages: Worked With vs. Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
#check missing data
missing_data = df.isnull()
print('\nValue counts for missing data in Employment column:\n')
print( missing_data['Employment'].value_counts())

print('\nValue counts for missing data in EdLevel column:\n')
print( missing_data['EdLevel'].value_counts())
```

```python
#check unique value
print('\nThere are', df['Employment'].nunique(), 'unique Employment values in the survey:')
print(df['Employment'].value_counts())

print('\nThere are', df['EdLevel'].nunique(), 'unique EdLevel values in the survey:')
print(df['EdLevel'].value_counts())
```

```python
#impute (replace) all the empty rows in the column EdLevel with the majority value
df['EdLevel'].replace(np.nan, "Bachelor’s degree (B.A., B.S., B.Eng., etc.)", inplace=True)
```

```python
#check unique value again
print('\nThere are', df['Employment'].nunique(), 'unique Employment values in the survey:')
print(df['Employment'].value_counts())

print('\nThere are', df['EdLevel'].nunique(), 'unique EdLevel values in the survey:')
print(df['EdLevel'].value_counts())
```

```python
#create cross-tab table
crosstab_counts = pd.crosstab(df['EdLevel'], df['Employment'])
crosstab_normalized = pd.crosstab(df['EdLevel'], df['Employment'], normalize='index') * 100 # In percentage

#display the frequency table
print("Frequency Cross-Tabulation (Counts):")
print(crosstab_counts)

print("\nNormalized Cross-Tabulation (Percentage within each EdLevel):")
print(crosstab_normalized.round(2)) # Rounding to 2 decimal places

#visualize the normalized data with a stacked bar plot
ax = crosstab_normalized.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Employment Status Distribution by Education Level (Percentages)')
plt.xlabel('Education Level')
plt.ylabel('Percentage (%)')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

#add percentage labels on the bars (optional, but helpful)
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    if height > 0:
        ax.text(x + width/2, y + height/2, f'{height:.1f}%', ha='center', va='center', fontsize=8, color='white')

plt.tight_layout()
plt.show()

#optional: Perform a Chi-square test to check for statistical significance
from scipy.stats import chi2_contingency

# The chi2_contingency function works with the frequency counts table
chi2, p_value, dof, expected = chi2_contingency(crosstab_counts)

print(f"\n--- Chi-Square Test Results ---")
print(f"Chi-square statistic: {chi2:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Result: The p-value is less than 0.05, suggesting a statistically significant relationship between Employment and Education Level (reject the null hypothesis of independence).")
else:
    print("Result: The p-value is greater than 0.05, suggesting no statistically significant relationship between Employment and Education Level (fail to reject the null hypothesis of independence).")
```
