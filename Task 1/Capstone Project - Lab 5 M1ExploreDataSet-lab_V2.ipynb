Print the mean age of the survey participants.


```python
## Write your code here
print('\nAge              Respondants')      
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df['Age'].value_counts())
```

```python
# The age ranges and the number of respondents in each category
age_data = {
    '25-34 years old': 23913,
    '35-44 years old': 14948,
    '18-24 years old': 14105,
    '45-54 years old': 6251,
    '55-64 years old': 2575,
    'Under 18 years old': 2570,
    '65 years or older': 773,
    'Prefer not to say': 322
}

# Define the midpoint for each age range for estimation
# For 'Under 18', we use an estimate of 17.5 years as an upper bound midpoint
# For '65 or older', we use 70 as a representative age
age_midpoints = {
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '18-24 years old': 21.0,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    'Under 18 years old': 17.5,
    '65 years or older': 70.0,
}

total_age_product_sum = 0
total_participants_for_mean = 0

# Iterate through the data, calculating the sum of ages and total participants
for age_range, count in age_data.items():
    if age_range in age_midpoints:
        midpoint = age_midpoints[age_range]
        total_age_product_sum += midpoint * count
        total_participants_for_mean += count

# Calculate the estimated mean age
estimated_mean_age = total_age_product_sum / total_participants_for_mean

print(f"Total participants used for calculation: {total_participants_for_mean}")
print(f"Total sum of age products: {total_age_product_sum}")
print(f"The estimated mean age of the participants is approximately: {estimated_mean_age:.2f} years old")
```
